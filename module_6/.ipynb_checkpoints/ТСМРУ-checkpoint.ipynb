```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler,OneHotEncoder
from sklearn.feature_selection import f_classif, mutual_info_classif
#метрики для оценки качества регрессии
from sklearn.metrics import recall_score, precision_score, f1_score,log_loss
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.metrics import precision_recall_curve, average_precision_score
from sklearn.metrics import confusion_matrix
```

```python
data=pd.read_csv('train.csv')
display(data)
data.info()
```

```python
# посмотрим на количество пропусков и их визульное распределение
display (data.isnull().sum())
plt.figure(figsize = (10,16))
ax=sns.heatmap(data.isnull(),cmap='rocket_r',cbar=False)
```

Как мы видим, пропуски есть только в графе **education**, попробуем присмотреться к ней подробнее. 

```python
# посмотрим на распределение образования
sns.countplot(data['education']) 
#видно, что большая часть заёмщиков имеет образование SCH
```

```python
# далее посмотрим на распределение данных для ячеек с незаполненным образованием:
empty_ed=data[data['education'].isnull()==True]
for column in empty_ed.columns:
  if column in ['client_id','education', 'income','default']:
    continue
  elif column in ['age','bki_request_cnt']:
    print(column.upper())
    display(empty_ed[column].value_counts(bins=5))
    print('='*50)
  else:
    print(column.upper())
    display(empty_ed[column].value_counts())
    print('='*50)
print('INCOME','\nMax:', empty_ed.income.max(),'\nMin:',empty_ed.income.min())
print('INCOME',empty_ed.income.value_counts(bins=11))
```

из интересных различий отметим:\
-возраст (от 30)\
-у большинства нет иномарок и хорошей работы\
-довольно существенную разницу в доходах, что говорит о том, что в группу попали не только люди с малым достатком\
-у большинства нет загран паспортов\
\
Как показало дальнейшее изучение на глаз нет четкой зависимости между наличием паспорта или автомобиля (в т.ч. иномарки) и уровнем образования, а что такое "хорошая работа" по заданию не определено, будем опираться на доход и возраст:

```python
display(data[(data['income'].between(5000, 95400))&((data['age']>30))]['education'].value_counts())
display(data[(data['income']>95400)&((data['age']>30))]['education'].value_counts())
```

Для людей с доходом **свыше** **95400** будем использовать **GRD**, а для людей с доходом **меньше** будем использовать **SCH**\
учитывая соотношение на графике мы скорее всего не сильно ошибемся. Заполним пропуски:

```python
data['education']=data.apply(lambda row: 'GRD' if pd.isnull(row['education']) and row['income']>=95400 else 'SCH' if pd.isnull(row['education']) and row['income']<95400 else row['education'],axis=1)
```

## Разделим колонки по типам данным для большего удобства работы и изучения

```python
# бинарные переменные
bin_cols=['sex','car','car_type','good_work','foreign_passport']
# категориальные переменные
cat_cols=['education','home_address','work_address']
# числовые переменные
num_cols=['age','decline_app_cnt','bki_request_cnt','income']
# целевую переменную рассмотрим отдельно
```

**2. рассмотрим бинарные переменные**

```python
# напишем функцию и сразу ее применим
def check_bin_cat(column):
  print('Уникальные значения для {a}:{b}'.format(a=col, b=data[column].unique()))
  display('Соотношение в %:',data[column].value_counts(normalize=True))
  plt.figure()
  sns.countplot(data[column])
  plt.show()
  print('='*80)

for col in bin_cols:
  check_bin_cat(col)
```

Стоит отметить:\
-соотношение полов почти одинаковое\
-у 70% заемщиков нет автомобиля\
-при этом у кого есть автомобиль, 80% обладают не иномаркой\
-у 83% процентов нет "хорошей" работы\
-почти у 85% нет загран паспорта

**3. рассмотрим категориальные переменные**

```python
for col in cat_cols:
  check_bin_cat(col)
```

Стоит отметить:\
-на долю SCH(53%) и GRD(31%) в образовании приходится 84% заёмщиков\
-home_address 3(2%) почти не встречается\
-в work_address преобладает 3 (61%),work_address - самая маленькая группа (11%)\
\
**3. рассмотрим числовые переменные**

```python
for col in num_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3-Q1
    plt.figure(figsize=(9, 9))
    sns.distplot(data[data['default'] == 0][col],
                 kde=False, color="g", label='default=0')
    sns.distplot(data[data['default'] == 1][col],
                 kde=False, color="r", label='default=1')
    sns.distplot(data[col], kde=False, label='default=0 & 1')
    plt.title(col)
    plt.legend()
    plt.show()
    print('для колонки:{}'.format(col), '\nГраницы выбросов:{}'.format(
        [Q1-1.5*IQR, Q3+1.5*IQR]), '\nКоличество выбросов:{}'.format(data[(data[col] < (Q1-1.5*IQR)) | (data[col] > (Q3+1.5*IQR))][col].count()))
    print('Минимальное значение:{a}, Максимальное значение:{b}'.format(
        a=data[col].min(), b=data[col].max()))
    print('='*70)
```

```python
# взглянем ближе на decline_app_cnt (количество отказанных прошлых заявок), так как по выбросам получены странные значения:
data['decline_app_cnt'].value_counts().plot(kind='bar',figsize=(10,6))
data[data['decline_app_cnt']>0]['default'].value_counts()
```

**Из всех графиков видно, что:**\
-это НЕ нормальное распределение\
-распределение смещено вправо\
-распределение внутри различных значений целевой переменной не меняется, а также имеет хвост справа\
-Присутствуют на первый взгляд выбросы:\
    - для income (хотя логика подсказывает, что доход может быть и больше 90000 у.е.)
    - для bki_request_cnt (хотя опять таки количество обращений в БКИ вполне может быть больше определенного значения)
    - для decline_app_cnt. Из графика видно, что подавляющее число заёмщиков не имеет прошлых отказов, поэтому любое значение отличное от 0, считается выбросом, хотя конечно это не так. 
-Таким образом пока от выбросов не будем избавляться, так как данные несут важную инфо для модели. Возможно данная позиция будет пересмотрена при построении модели\
\
Посмотрим **отличается ли распределения** для дефолтных и недефолтных клиентов

```python
def boxplot(col):
    fig, ax = plt.subplots(figsize=(8,8))
    sns.boxplot(x='default', y=col, data=data, ax=ax)
    ax.set_title('Boxplot для '+col)
    plt.show()

for col in num_cols:
    boxplot(col)
```

Исходя из боксплотов можно сделать выводы, что:\
-Дефолтные клиенты в среднем младше,\
-Дефолтные клиенты имеют в среднем большее количество запросов в БКИ и отказов,\
-Дефолтные клиенты имеют в среднем более низкий доход.\
Предварительно можно сказать, что все эти признаки следует использовать для построения модели. Дополнительно проверим на следующем этапе\
\
Также посмотрим, что будет, если **логарифмировать** данные

```python
data1 = data.copy()
for col in num_cols:
    data1[col]=data1[col].apply(lambda x:np.log(x+1))
    plt.figure(figsize=(9, 9))
    sns.distplot(data1[data1['default'] == 0][col],
                 kde=False, color="g", label='default=0')
    sns.distplot(data1[data1['default'] == 1][col],
                 kde=False, color="r", label='default=1')
    sns.distplot(data1[col], kde=False, label='default=0 & 1')
    plt.title(col)
    plt.legend()
    plt.show()
    print('='*70)
```

Для признаков **age** и **income** распределение стало выглядеть **более нормальным**.\
возможно воспользуемся этим при построении модели

## Отбор признаков для модели

посмотрим на корреляцию

```python
# построим две корреляции для признаков
# Пирсона
pirson=data[num_cols+['default']].corr()
plt.figure(figsize = (10,10))
sns.heatmap(pirson,cmap='rocket_r',annot=True)
```

```python
# Спирмана, чтобы лучше просмотреть нечисловые признаки
spearman=data[set(data.columns)-set(num_cols)-set(['client_id'])].corr(method='spearman')
plt.figure(figsize = (12,10))
sns.heatmap(spearman,cmap='rocket_r',annot=True)
```

**Выводы:**
- отметим, что для с default на первый взгляд ни у одной переменной нет сильной линейной связи
- Стоит отметить **наличие линейной связи** между:
  - education_SCH и education_GRD (-0.72)
  - home_address и work_address (0.79)
  - car и car_type (0.7)
Это не прямая линейная зависимость, но возможно при построении модели возникнут сложности. Пока просто зафиксируем данные выводы

\
Теперь проведем **однофакторный дисперсионный анализ (ANOVA)** для числовых переменных\
Также проведем **оценку энтропии в группах категориальных переменных**, т.е. воспользуемся функцией mutual_info_classif

```python
#ANOVA
imp_num=pd.Series(f_classif(data[num_cols],data['default'])[0],index=num_cols)
imp_num.sort_values(inplace=True)
imp_num.plot(kind='barh')
```

```python
#mutual_info_classif
imp_cat=pd.Series(mutual_info_classif(data[set(data.columns)-set(num_cols)-set(['client_id','default'])],data['default'],discrete_features=True),index=set(data.columns)-set(num_cols)-set(['client_id','default']))
imp_cat.sort_values(inplace=True)
imp_cat.plot(kind='barh',figsize=(8,8))
```

**Данная информация о важности признаков будет полезна при создании модели и работы с признаками.** Пока сохраним её\
\
Также стоит отметить, что от одной колонки из созданных **education** нужно будет **избавиться**, чтобы решить проблему **мультиколлинеарности**
