```python
# напишем функцию для подсчета и отображения всех интересующих метрик качества
def all_metrics(y_tr_test, y_pr_test, proba_test, y_tr_valid, y_pr_valid, proba_valid):
    fpr_test, tpr_test, threshold_test = roc_curve(y_tr_test, proba_test)
    roc_auc_test = roc_auc_score(y_tr_test, proba_test)

    fpr_valid, tpr_valid, threshold_valid = roc_curve(y_tr_valid, proba_valid)
    roc_auc_valid = roc_auc_score(y_tr_valid, proba_valid)

    plt.figure(figsize=(10, 8))
    plt.plot([0, 1], label='Baseline', linestyle='--')
    plt.plot(fpr_test, tpr_test, label='TRAIN PART')
    plt.plot(fpr_valid, tpr_valid, label='VALID PART')
    plt.title('Logistic Regression ROC AUC \nfor TRAIN PART ={a}\nfor VALID PART ={b}'.format(
        a=round(roc_auc_test, 4), b=round(roc_auc_valid, 4)))
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.legend(loc='lower right')

    precision_test, recall_test, thresholds_test = precision_recall_curve(
        y_tr_test, proba_test)
    av_prec_sc_test = average_precision_score(y_tr_test, proba_test)

    precision_valid, recall_valid, thresholds_valid = precision_recall_curve(
        y_tr_valid, proba_valid)
    av_prec_sc_valid = average_precision_score(y_tr_valid, proba_valid)

    plt.figure(figsize=(10, 8))
    plt.plot(recall_test, precision_test, label='TRAIN PART')
    plt.plot(recall_valid, precision_valid, label='VALID PART')
    plt.title('Average Precision for TRAIN PART={c}\nAverage Precision for for VALID PART={d}'.format(
        c=round(av_prec_sc_test, 4), d=round(av_prec_sc_valid, 4)))
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.legend(loc='lower right')
    plt.show()

    print('='*70)
    print(' TRAIN PART\n Log loss: {a}\n F_1: {b}\n Precision: {c}\n Recall: {d}'.format(a=round(log_loss(y_tr_test, proba_test), 5), b=round(
        f1_score(y_tr_test, y_pr_test), 5), c=precision_score(y_tr_test, y_pr_test), d=recall_score(y_tr_test, y_pr_test)))
    print('-'*70)
    tn_test, fp_test, fn_test, tp_test = confusion_matrix(
        y_tr_test, y_pr_test).ravel()
    print(' TP:', tp_test, 'FP:', fp_test)
    print(' FN:', fn_test, 'TN:', tn_test)

    print('='*70)
    print(' VALID PART\n Log loss: {a}\n F_1: {b}\n Precision: {c}\n Recall: {d}'.format(a=round(log_loss(y_tr_valid, proba_valid), 5), b=round(
        f1_score(y_tr_valid, y_pr_valid), 5), c=precision_score(y_tr_valid, y_pr_valid), d=recall_score(y_tr_valid, y_pr_valid)))
    print('-'*70)
    tn_valid, fp_valid, fn_valid, tp_valid = confusion_matrix(
        y_tr_valid, y_pr_valid).ravel()
    print(' TP:', tp_valid, 'FP:', fp_valid)
    print(' FN:', fn_valid, 'TN:', tn_valid)
```

## Построение моделей

### Первая модель
построим модель по имеющимся данным без дополнительных обработок и оценим ее\
поскольку для тестовых данных у нас нет значение целевой переменной, будем работать только с трейн (train.csv), разделив ее на тестовую часть и валидационную


```python
df_test = data[data['train'] == 1].copy()
df_test.drop(['client_id', 'app_date', 'train'], axis=1, inplace=True)
display(df_test.head(2))
df_test.info()
```

```python
# стандартизируем
scaler = StandardScaler()
scaler.fit(df_test[num_cols])  # только по train.csv части
df_test[num_cols] = scaler.transform(df_test[num_cols])
```

```python
# задаем значения X и y стратифицированным разбиением (показало себя лучше):
sss = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
train_indices, valid_indices = [split for split in sss.split(
    df_test[set(df_test.columns)-set(['default'])], df_test['default'])][0]
# разобьем выборки
X_train = df_test.iloc[train_indices][set(
    df_test.columns)-set(['default'])].values
X_valid = df_test.iloc[valid_indices][set(
    df_test.columns)-set(['default'])].values
y_train = df_test.iloc[train_indices]['default'].values
y_valid = df_test.iloc[valid_indices]['default'].values
```

```python
# строим модель, выводим метрики
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
y_train_pred = model.predict(X_train)
y_train_probs = model.predict_proba(X_train)[:, 1]
y_valid_pred = model.predict(X_valid)
y_valid_probs = model.predict_proba(X_valid)[:, 1]
all_metrics(y_train, y_train_pred, y_train_probs,
            y_valid, y_valid_pred, y_valid_probs)
```

**пока очевидны две проблемы**
- модель склоняется к более популярному классу (из за несбалансированности выборки) - см FN
- модель требует задавать значение max_iter больше дефолтного. это возникает (см.ниже) из-за вероятно из-за decline_app_cnt, bki_request_cnt, income
